### Bài 3:

```python
import math
a, b, c = map(float, input("Nhập ba cạnh a, b, c: ").split())
if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        print("Tam giác đều")
    elif a == b or b == c or a == c:
        print("Tam giác cân")
    elif a*a + b*b == c*c or a*a + c*c == b*b or b*b + c*c == a*a:
        print("Tam giác vuông")
    else:
        print("Tam giác thường")
    P = a + b + c
    p = P / 2
    S = math.sqrt(p * (p - a) * (p - b) * (p - c))
    print("Chu vi:", P)
    print("Diện tích:", S)
else:
    print("Không là ba cạnh của một tam giác")
```

### Bài 5:

```python
import math
a = float(input("Nhập hệ số a: "))
b = float(input("Nhập hệ số b: "))
c = float(input("Nhập hệ số c: "))
if a == 0:
    if b == 0:
        if c == 0:
            print("Phương trình có vô số nghiệm")
        else:
            print("Phương trình vô nghiệm")
    else:
        print("Phương trình có nghiệm x =", -c/b)
else:
    delta = b**2 - 4*a*c
    if delta < 0:
        print("Phương trình vô nghiệm")
    elif delta == 0:
        x = -b / (2*a)
        print("Phương trình có nghiệm kép x =", x)
    else:
        x1 = (-b - math.sqrt(delta)) / (2*a)
        x2 = (-b + math.sqrt(delta)) / (2*a)
        print("Phương trình có 2 nghiệm phân biệt:")
        print("x1 =", x1)
        print("x2 =", x2)
```

### Bài 8:

```python
a1, b1, c1 = map(float, input().split())
a2, b2, c2 = map(float, input().split())
D = a1 * b2 - a2 * b1
Dx = c1 * b2 - c2 * b1
Dy = a1 * c2 - a2 * c1
if D == 0:
    if Dx == 0 and Dy == 0:
        print("Vô số nghiệm")
    else:
        print("Vô nghiệm")
else:
    x = Dx / D
    y = Dy / D
    print(f"{x:.2f} {y:.2f}")
```

### Bài 10:

```python
x0, y0, r = map(float, input("Nhập tâm (x0, y0) và bán kính r: ").split())
x, y = map(float, input("Nhập tọa độ điểm A(x, y): ").split())
d2 = (x - x0)**2 + (y - y0)**2
if d2 == r**2:
    print("Điểm A nằm trên đường tròn")
elif d2 < r**2:
    print("Điểm A nằm trong đường tròn")
else:
    print("Điểm A nằm ngoài đường tròn")
```

```python
x = float(input())
y = float(input())
if x == 0:
    if y > 0:
        print("0.00")
    elif y == 0:
        print("1.00")
    else:
        print("Không xác định")
elif x < 0 and y != int(y):
    print("Không xác định") 
else:
    ket_qua = x ** y
    print(f"{ket_qua:.2f}")
```
