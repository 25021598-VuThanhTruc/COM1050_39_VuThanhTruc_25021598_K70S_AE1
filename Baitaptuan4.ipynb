### W4A6

```python
import math
def day(n):
    if n < 2:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True
while True:
    try:
        a, b = map(int, input("Nhập 2 số a, b (a <= b): ")).split()
        if a < 0 or b < 0:
            print("a, b không thể là số âm")
            continue
        if a > b:
            print("Chưa thỏa mãn điều kiện a <= b")
            continue
        tong = 0
        for i in range(a, b+1):
            if day(i):
                tong += i
        print(f"Tổng các số nguyên tố từ {a} đến {b} là {tong}")
    except:
        break
```

### W4A7

```python
def uoc_lon_nhat(n):
    largest = -1
    while n % 2 == 0:
        largest = 2
        n //= 2
    i = 3
    while i*i <= n:
        while n % i == 0:
            largest = i
            n //= i
        i += 2
    if n > 2:
        largest = n
    return largest
n = int(input("Nhập vào số dương n (n >= 2): "))
print("Ước nguyên tố lớn nhất của ", n ,"là ", uoc_lon_nhat(n))
```

### W4A8

```python
def reserve_numbers(num):
    return int(str(num)[::-1])
def is_palindrome(num):
    return str(num) == str(num)[::-1]
n = int(input("Nhập số nguyên dương n: "))
step = 0
while not is_palindrome(n):
    n += reserve_numbers(n)
    step += 1
print(f"Số bước là {step}")
print(f"Palindrome là {n}")
```

### W4A9

```python
import math
def khac_nhau(num):
    s = str(num)
    return len(s) == len(set(s))
n = int(input("Nhập số nguyên dương n: "))
result = []
for i in range(2, int(math.sqrt(n)) + 1):
    chinh_phuong = i * i
    if khac_nhau(chinh_phuong):
        result.append(chinh_phuong)
if result:
    print(*result)
else:
    print("(no number)")
```

### W4A10

```python
def do_dai_collatz(m):
    dem = 1
    while m != 1:
        if m % 2 == 0:
            m //= 2
        else:
            m = 3 * m + 1
        dem += 1
    return dem
n = int(input("Nhập số nguyên dương n: "))
so_tot = 1
do_dai_max = do_dai_collatz(1)
for x in range(2, n + 1):
    do_dai = do_dai_collatz(x)
    if do_dai > do_dai_max or (do_dai == do_dai_max and x < so_tot):
        so_tot = x
        do_dai_max = do_dai
print(so_tot, do_dai_max)
```

### W4A11

```python
def uoc_chan(num):
    so_luong = 0
    for i in range(2, num + 1, 2):
        if num % 1 == 0:
            so_luong +=1
    return so_luong
n = int(input("Nhập số nguyên dương n (n<1000000): "))
while n > 10**6 or n <= 0:
    n = int(input("Nhập số nguyên dương n (0<n<1000000): "))
print("Số lượng ước chẵn của ", n, "là", uoc_chan(n))
```

```python
def tong_uoc(num):
    tong = 0
    for i in range(1, num):
        if num % i == 0:
            tong += i
    return tong
a, b = map(int, input("Nhập 2 số a và b cách nhau bởi dấu cách: ").split())
if tong_uoc(a) == b and tong_uoc(b) == a:
    print("True")
else:
    print("False")
```
