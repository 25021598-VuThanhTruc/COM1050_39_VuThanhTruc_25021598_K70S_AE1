### Bài 4

```python
h = int(input("Nhập chiều cao tam giác: "))
for i in range(1, h + 1):
    khoang_trang = h - i
    if i == 1:
        print(" " * khoang_trang + "*")
    elif i == h:
        print("*" * (2 * h - 1))
    else:
        print(" " * khoang_trang + "*" + " " * (2 * i - 3) + "*")
```

### Bài 14

```python
print("Nhập dãy số nguyên, nhập -1 để kết thúc:")
min_val = None
max_val = None
while True:
    x = int(input())
    if x == -1:
        break
    if min_val is None and max_val is None:
        min_val = x
        max_val = x
    else:
        if x < min_val:
            min_val = x
        if x > max_val:
            max_val = x
if min_val is None:
    print("Không có số nào được nhập!")
else:
    print("Số nhỏ nhất:", min_val)
    print("Số lớn nhất:", max_val)
```

### Bài 16

```python
n = int(input("Nhập số nguyên dương n:"))
chan = 0
le = 0
while n > 0:
    so = n % 10
    if so % 2 == 0:
        chan += 1
    else:
        le += 1
    n //= 10
print("Số chữ số chẵn là", chan)
print("Số chữ số lẻ là", le)
```

### Bài 18

```python
n = int(input("Nhập số nguyên dương n: "))
while n < 1:
    n = int(input("n không hợp lệ, vui lòng nhập lại"))
result = False
for k in range(1, int(n**(1/3)) + 2):
    if 3**k == n:
        result = True
        break
if result:
    print("Có")
else:
    print("Không")
```

### Bài 19

```python
A = int(input("Nhập số nguyên dương A: "))
while A < 1:
    A = int(input("A không hợp lệ, vui lòng nhập lại: "))
f1 = 1
f2 = 1
f_max = f1
while f2 <= A:
    f_max = f2
    f_next = f1 + f2
    f1 = f2
    f2 = f_next
print("Số Fibonacci lớn nhất không vượt quá", A, "là:", f_max)
```
