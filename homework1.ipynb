### Bài 2:

```python
a = float(input("Nhập chiều rộng a: "))
b = float(input("Nhập chiều dài b: "))
pi = 3.14
S_hcn = a * b
d = min(a, b)
r = d / 2
S_tron = pi * r ** 2
S = S_hcn - S_tron
print(f"Diện tích phần trồng cây: {S:.2f}")
```

### Bài 6:

```python
import math
a, b, c = map(float, input("Nhập 3 số a, b, c: ").split())
if a + b > c and a + c > b and b + c > a:
    p = (a + b + c) / 2
    S = math.sqrt(p * (p - a) * (p - b) * (p - c))
    print(f"Diện tích tam giác: {S:.1f}")
else:
    print("Khong phai 3 canh cua tam giac")
```

### Bài 8:

```python
ten = input("Nhập tên chủ hộ: ")
so_cu = int(input("Nhập chỉ số tháng trước: "))
so_moi = int(input("Nhập chỉ số tháng này: "))
so_dien = so_moi - so_cu
tien = 0
if so_dien <= 50:
    tien = so_dien * 1680
elif so_dien <= 100:
    tien = 50 * 1680 + (so_dien - 50) * 1800
elif so_dien <= 200:
    tien = 50 * 1680 + 50 * 1800 + (so_dien - 100) * 2080
elif so_dien <= 300:
    tien = 50 * 1680 + 50 * 1800 + 100 * 2080 + (so_dien - 200) * 2380
elif so_dien <= 400:
    tien = 50 * 1680 + 50 * 1800 + 100 * 2080 + 100 * 2380 + (so_dien - 300) * 2700
else:
    tien = (50 * 1680 + 50 * 1800 + 100 * 2080 + 100 * 2380 +
            100 * 2700 + (so_dien - 400) * 3050)
vat = round(tien * 0.08)
tong_tien = tien + vat
print("Họ và tên:", ten)
print("Tiền phải trả là:", tong_tien)
```

### Bài 9:

```python
n = int(input("Nhập một số nguyên từ 0 đến 9: "))
match n:
    case 0:
        print("Không")
    case 1:
        print("Một")
    case 2:
        print("Hai")
    case 3:
        print("Ba")
    case 4:
        print("Bốn")
    case 5:
        print("Năm")
    case 6:
        print("Sáu")
    case 7:
        print("Bảy")
    case 8:
        print("Tám")
    case 9:
        print("Chín")
    case _:
        print("Số không hợp lệ, phải từ 0 đến 9")
```
