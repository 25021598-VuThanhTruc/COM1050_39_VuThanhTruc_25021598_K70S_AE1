### W5A1

```python
def Max_Of_Two(a: int, b: int) -> int:
    if a > b:
        return a
    else:
        return b
x, y = map(int, input("Nhập a, b").split())
print(Max_Of_Two(x, y))
```

### W5A2

```python
from typing import Tuple
def swap(a: int, b: int) -> Tuple[int, int]:
    return b, a
x, y = map(int, input("Nhập a, b").split())
print("Sau khi hoán đổi: ", swap(x, y))
```

### W5A3

```python
def nguyen_to(n: int) -> bool:
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
x = int(input())
print(nguyen_to(x))
```

### W5A4

```python
def perfect_number(n: int) -> bool:
    if n < 2:
        return False
    tong = 0
    for i in range(1, n):
        if n % i == 0:
            tong += i
    return tong == n
x = int(input())
print(perfect_number(x))
```

### W5A5

```python
from typing import List
def find(nums: List[int], k: int) -> int:
    for i in range(len(nums)):
        if nums[i] == k:
            return i
    return -1
nums = list(map(int, input().split()))
k = int(input())
index = find(nums, k)
if index != -1:
    print(f"Phần tử {k} ở vị trí {index + 1}")
else:
    print(f"Phần tử {k} không tồn tại trong danh sách")
```

### W5A6

```python
def giai_thua(n: int) -> int:
    for i in range(1, n):
        n *= i
    return n
x = int(input())
while x < 1:
    x = int(input())
print(giai_thua(x))
```

### W5A7

```python
def calculator(num1: float, num2: float, operat: str) -> float:
    if operat == '+':
        return num1 + num2
    elif operat == '-':
        return num1 - num2
    elif operat == '*':
        return num1 * num2
    elif operat == '/':
        if num2 == 0:
            return "Lỗi: Không thể chia cho 0"
        return num1 / num2
    else:
        return "Lỗi: Phép toán không hợp lệ"
a, b = map(float, input("Nhập a, b").split())
c = input("Nhập phép toán")
print(calculator(a, b, c))
```

### W5A8

```python
def hamming(x, y):
    xor_result = x ^ y
    khac = 0
    while xor_result:
        khac += 1
        xor_result &= xor_result - 1
    return khac
a, b = map(int, input().split())
print(hamming(a, b))
```

### W5A9

```python
def tong_chu_so(n):
    tong = 0
    while n > 0:
        tong += n % 10
        n //= 10
    return tong
x = int(input())
while x <= 0:
    x = int(input())
print(tong_chu_so(x))
```

### W5A10

```python
def is_isomorphic(a: str, b: str) -> bool:
    if len(a) != len(b):
        return False
    map_a_to_b = {}
    map_b_to_a = {}
    for char_a, char_b in zip(a, b):
        if char_a in map_a_to_b:
            if map_a_to_b[char_a] != char_b:
                return False
        else:
            map_a_to_b[char_a] = char_b
        if char_b in map_b_to_a:
            if map_b_to_a[char_b] != char_a:
                return False
        else:
            map_b_to_a[char_b] = char_a
    return True
word1 = input()
word2 = input()
if is_isomorphic(word1, word2):
    print(f"'{word1}' và '{word2}' là đẳng cấu")
else:
    print(f"'{word1}' và '{word2}' không là đẳng cấu")
```
