##### W2A6

```python
chu = input("Nhập một chữ cái thường từ 'a' đến 'z':")
if len(chu) == 1 and 'a' <= chu <= 'z':
    ma_unicode = ord(chu)
    print("Mã Unicode của chữ cái là: ", ma_unicode)
    chuhoa = chr(ma_unicode - 32)
    print("Chữ hoa của chữ cái là: ", chuhoa)
else:
    print("Vui lòng nhập đúng một chữ cái từ 'a' đến 'z':")
```
